```python
x = [1,2,3]
y = [1,4,9]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('Sales')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.plot(x,y)
plt.title('Sales Performance')
plt.xlabel('Sales')
plt.ylabel('Revenue')
plt.show()
```

```python
z = [10,5,0]
plt.plot(x,y)
plt.plot(x,z)
plt.title('Sales Performance')
plt.xlabel('Sales')
plt.ylabel('Revenue')
plt.show()
```
